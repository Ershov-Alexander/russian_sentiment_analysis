```
from src.evaluation import evaluate
from transformers import (
    BertTokenizer,
    BertForSequenceClassification
)

import pandas as pd

from time import time
```

```
model = BertForSequenceClassification.from_pretrained("/content/gdrive/My Drive/russian_sentiment_analysis/models/model2/")
```

```
tokenizer = BertTokenizer.from_pretrained("DeepPavlov/rubert-base-cased-conversational")
```

```
test = pd.read_csv("/content/gdrive/My Drive/russian_sentiment_analysis/data/preprocessed_test.csv", sep="&")
```

```
t0 = time()

evaluate(model, tokenizer, data=test, max_length=192)

print(f"\nEvaluation took {int(time() - t0)} seconds")
```
